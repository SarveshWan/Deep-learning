**step 1:Read the Dataset**

```python
import pandas as pd
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/booking_train.csv"
df = pd.read_csv(url1)
```

```python
df.head()
```

Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

step 3:Seperate X and Y (booking_status)**bold text** bold text

```python
X = df.drop(columns=['id','booking_status'])
Y = df[['booking_status']]
X.head()
```

```python
Y.head()
```

**Step 4: Perform preprocessing on X**

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps =[('impute',SimpleImputer(strategy = 'median')),
                            ('scaler',StandardScaler())]).set_output(transform = 'pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

**Step : 5 Train test split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test = train_test_split(X_pre,Y,test_size=0.2,random_state=23)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

Dropout layer- Randomly drop some%of neuron links


```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```python
# Initialize the model
model = Sequential()
# Add the input layer and the first hidden layer
model.add(Input(shape=(X_train.shape[1],)))

# Hidden layer 1
model.add(Dense(64,activation='relu'))

# add Dropout layer 1
model.add(Dropout(0.4))

# Hidden layer 2
model.add(Dense(32,activation='relu'))
# add Dropout layer 2
model.add(Dropout(0.3))
# Add Outer layer
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train, epochs=60, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(hist.history['loss'],label = 'Training Loss')
plt.plot(hist.history['val_loss'],label = 'Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary cross entropy Loss')
plt.title('Learning curve')
plt.legend()
plt.show()
```

```python
model.evaluate(X_train, Y_train)
```

```python
model.evaluate(X_test,Y_test)
```

**Get predictions on train and test**

```python
yprob_train = model.predict(X_train)
yprob_test = model.predict(X_test)
```
